8) Perform the data cleaning operations mentioned so far in class

a) Delete the column education and the number of open complaints from the dataframe.

b)Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

c)Check for duplicate rows in the data and remove if any.

```python
new_df.columns
```

```python
# a) Delete Columns
data_drop = new_df.drop(["education","number_of_open_complaints"], axis=1)
data_drop.head()
```

```python
# Column Correction 
data_drop['customer_lifetime_value'] = data_drop['customer_lifetime_value'].fillna(0) # change Na to 0
data_1 = data_drop
data_1.head()
```

```python
np.dtype(data_1['customer_lifetime_value'])
```

```python
data_1['customer_lifetime_value'] = data_1['customer_lifetime_value'].astype(str) # converts object into string
data_1['customer_lifetime_value'] = data_1['customer_lifetime_value'].str.replace(r'%', '') # remove % sign
data_1.head()
```

```python
data_1['customer_lifetime_value'] = data_1['customer_lifetime_value'].astype(float) # converts string into float
data_1.head()
```

```python
np.dtype(data_1['customer_lifetime_value'])
```

```python
data_1['customer_lifetime_value'] = round(data_1['customer_lifetime_value'],0) # round to the nearest decimal 
data_1.head()
```

```python
data_1['customer_lifetime_value'] = data_1['customer_lifetime_value'].astype(int) # converts the float into an integer
np.dtype(data_1['customer_lifetime_value'])
```

```python
# multiply a column by a value
data_1['customer_lifetime_value']  = data_1['customer_lifetime_value'] *100
data_1.head()
```

```python
# check for duplicates
duplicates = data_1[data_1.duplicated()]
duplicates.head()
```

```python
# delete duplicates
duplicates = duplicates.drop_duplicates()
duplicates.head()
```

```python
# filter out income that is lower than 0
data_1 = data_1[data_1['income'] > 0] 
data_1.head()
```
